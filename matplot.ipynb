###  MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[2,3,4,5]
plt.plot(x,y,marker='x',color='green',linestyle='-.')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("Line plot")
```

```python
plt.plot([1,2,3,4,5],[2,3,3,5,3])
```

```python
marks=[56,36,78,24,34]
division=["A","B","C","D","E"]
plt.bar(marks,division,width=3,color="green",label="Marks")
plt.legend(loc="best")
```

```python
plt.barh(marks,division,color="green",label="Marks")
```

```python
lang=['Telugu','Hindi','Tamil']
rating=[45,34,56]
plt.pie(rating,labels=lang,autopct="%1.1f%%",startangle=40,shadow=True,labeldistance=1.5)
```
